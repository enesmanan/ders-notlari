## Lineer Denklem Sistemleri
Birden fazla doğrusal denklemin bir arada bulunduğu sistemlerdir. Doğrusal denklemler ise değişkenlerin en fazla birinci dereceden olabileceği denklemlerdir. Doğrusal denklem sistemlerinin çözümü için grafik yöntemi, yerine koyma yöntemi, eşitlik yöntemi veya matris yöntemi gibi farklı yöntemler kullanılabilir.

Örnek:  
$2x + y = 5$

$x - y = 1$

## Lineer Denklem Sistemlerinin Çözüm Durumları

Lineer Denklem Sistemlerinin Çözüm Durumları, denklem sistemini sağlayan sıralı çiftlerin sayısına göre üç farklı şekilde olabilir:

### ***1. Durum: Tek bir çözüm olabilir.***
Denklem sistemini sağlayan tek bir sıralı çift vardır. Bu durumda, denklem sistemini grafikle çözmek istediğimizde iki doğrunun bir noktada kesiştiğini görürüz. Bu nokta, denklem sisteminin tek çözümüdür. 

<b>Örnek:</b>

$x+y=5$

$x−y=1$
denklem sisteminin çözümünde $x=3$ ve $y=2$ olacaktır. 

### Python İle Grafiksel Gösterimi

```python
import matplotlib.pyplot as plt

# Denklemleri y = mx + n şeklinde yazalım
m1 = -1 # İlk denklemin eğimi
n1 = 5 # İlk denklemin sabiti
m2 = 1 # İkinci denklemin eğimi
n2 = 1 # İkinci denklemin sabiti

# x değerleri için bir aralık belirleyelim
x = range(-10, 10)

# Her x değeri için y değerlerini hesaplayalım
y1 = [m1 * i + n1 for i in x] # İlk denklemin y değerleri
y2 = [m2 * i + n2 for i in x] # İkinci denklemin y değerleri

# Grafikte doğruları çizdirelim
plt.plot(x, y1, color="blue", label="x + y = 5") # İlk denklemi mavi renkte çizdiriyoruz
plt.plot(x, y2, color="red", label="x - y = 1") # İkinci denklemi kırmızı renkte çizdiriyoruz

# Grafikte başlık, eksen isimleri ve açıklama ekleyelim
plt.title("Tek Çözümlü Denklem Sistemi")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Grafikte kesişme noktasını işaretleyelim
x_intersect = 2
y_intersect = 3
plt.scatter(x_intersect, y_intersect, color="black", marker="o") # Kesişme noktasını siyah bir daire ile işaretleyelim
plt.annotate(f"({x_intersect}, {y_intersect})", (x_intersect, y_intersect), xytext=(x_intersect+1, y_intersect+1)) # Kesişme noktasının koordinatlarını yazalım

# Grafiği gösterelim
plt.show()
```

### ***2. Durum: Çözümsüz olabilir.*** 
Denklem sistemini sağlayan hiçbir sıralı çift yoktur. Bu durumda, denklem sistemini grafikle çözmek istediğimizde iki doğrunun paralel olduğunu ve hiçbir noktada kesişmediğini görürüz.

<b>Örnek:</b> 

$2x+3y=5$

$4x+6y=7$

### Python İle Grafiksel Gösterimi

```python
import matplotlib.pyplot as plt

# Denklemleri y = mx + n şeklinde yazalım
m1 = 2/3 # İlk denklemin eğimi
n1 = 5/3 # İlk denklemin sabiti
m2 = 2/3 # İkinci denklemin eğimi
n2 = 7/6 # İkinci denklemin sabiti

# x değerleri için bir aralık belirleyelim
x = range(-10, 10)

# Her x değeri için y değerlerini hesaplayalım
y1 = [m1 * i + n1 for i in x] # İlk denklemin y değerleri
y2 = [m2 * i + n2 for i in x] # İkinci denklemin y değerleri

# Grafikte doğruları çizdirelim
plt.plot(x, y1, color="blue", label="2x + 3y = 5") # İlk denklemi mavi renkte çizdiriyoruz
plt.plot(x, y2, color="red", label="4x + 6y = 7") # İkinci denklemi kırmızı renkte çizdiriyoruz

# Grafikte başlık, eksen isimleri ve açıklama ekleyelim
plt.title("Çözümsüz Denklem Sistemi")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Grafiği gösterelim
plt.show()
```

Grafiğe baktığımız zaman bu denklemin doğrudaki her noktada çözümü vardır.

## Lineer Denklem Sistemlerinin Matrisler İle Gösterimi

Denklem sisteminin bileşenlerini (değişkenler, katsayılar ve sabit terimler) matris formunda getirmektir. 

<b>Örnek:</b>

$2x + 3y = 5$

$x - y = 2$
denklem sisteminin matris ile gösterimi şöyle olabilir:

Değişkenler Matrisi = $\begin{bmatrix} x \\ y \end{bmatrix}$

Katsayılar Matrisi = $\begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix}$

Sabit Terimler Matrisi = $\begin{bmatrix} 5 \\ 2 \end{bmatrix}$

Matris Denklemi = $\begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 5 \\ 2 \end{bmatrix}$

Artırılmış matris: Denklem sisteminin sabit terimlerini katsayılar matrisinin sonuna ekleyip, bir artırılmış matris oluşturabiliriz. 

Artırılmış Matris = $\begin{bmatrix} 2 & 3 & | & 5 \\ 1 & -1 & | & 2 \end{bmatrix}$

Dikey çizgi sabit terimleri ayırmak için kullanılır.

<hr style="border: 0.5px solid gray; margin: 25px 0;">

##  $AX=b$ Metodu

$AX=b$ metodu, lineer denklem sistemlerinde matris denklemi şeklinde yazılan bir denklem sisteminin çözümü için kullanılan bir yöntemdir . Bu yöntemde, $A$ katsayılar matrisi, $X$ değişkenler matrisi ve $b$ sabit terimler matrisi olmak üzere, $AX=b$ şeklinde yazılan bir matris denkleminin çözümü için $X=A^{-1}b$ formülü kullanılır. Burada $A^{-1}$, $A$ matrisinin tersi olup, $AA^{-1}=I$ eşitliğini sağlayan bir matristir. $I$, birim matris olarak adlandırılan ve köşegeninde 1, diğer elemanlarında 0 olan bir matristir. 

<b>Örnek:</b>

$\begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 5 \\ 2 \end{bmatrix}$

şeklinde yazılan bir matris denkleminin çözümü için:

$\begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix}^{-1} \begin{bmatrix} 5 \\ 2 \end{bmatrix}$

formülü kullanılır. Burada:

$\begin{bmatrix} 2 & 3 \\ 1 & -1 \end{bmatrix}^{-1} = \frac{1}{-5}\begin{bmatrix} -1 & -3 \\ -1 & 2 \end{bmatrix} = \begin{bmatrix} 0.2 & 0.6 \\ 0.2 & -0.4 \end{bmatrix}$

matrisinin tersidir. Böylece:

$\begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0.2 & 0.6 \\ 0.2 & -0.4 \end{bmatrix}\begin{bmatrix} 5 \\ 2 \end{bmatrix}= \begin{bmatrix} (0.2)(5) + (0.6)(2) \\ (0.2)(5) + (-0.4)(2) \end{bmatrix}= \begin{bmatrix} 2.2 \\ 0.6 \end{bmatrix}$

olur. Bu da demektir ki:

$x=2.2$
$y=0.6$

<b>Teorem:</b> Bir matrisin rankı, o matrisin doğrusal bağımsız saıt sayısını verir.

<b>Teorem:</b> $AX=b$ lineer denklem sisteminin çözümlü olabilmesi için gerek ve yeter şart $\to$ $rank(A)=rank(A|b)$

$*$ $A^{-1}$'in varlığı ancak $|A|≠0$ ise mümkündür. Yani $A$ tekil değil ise $A^{-1}$ vardır.

<b>Örnek:</b>

$x-y=4$

$x-y=0$

Denklem sisteminin çözüm kümesi var mıdır?

<b>Çözüm:</b>

Öncelikle, katsayılar matrisi $A$ ve sabitler matrisi $b$'yi yazalım.

$
A=\begin{bmatrix}
1 & -1 \\
1 & -1
\end{bmatrix}
$

$
b=\begin{bmatrix}
4 \\
0
\end{bmatrix}
$

Artırılmış matris $A|b$ ise şöyle olur:

$
A|b=\begin{bmatrix}
1 & -1 & | & 4 \\
1 & -1 & | & 0
\end{bmatrix}
$

Şimdi, $A$ matrisinin determinantını hesaplayalım.

$
|A|=\begin{vmatrix}
1 & -1 \\
1 & -1
\end{vmatrix}=1\times (-1)-(-1)\times 1=0
$

Görüldüğü gibi, $|A|=0$ olduğundan, $A$ matrisinin tersi yoktur. Yani $A^{-1}$ mümkün değildir. Bu durumda, denklem sistemi **Cramer olmayan** bir denklem sistemidir.

Bu durumda, denklem sisteminin çözümlü olabilmesi için gerek ve yeter şart olan $rank(A)=rank(A|b)$ eşitliğini kontrol etmemiz gerekir.

Bunun için, $A$ ve $A|b$ matrislerinin ranklarını bulalım.

Bir matrisin rankı, o matrisin doğrusal bağımsız satır sayısını verir. Doğrusal bağımsız satır sayısı ise, o satırların birbirine oranlanamayacağı anlamına gelir.

$A$ matrisine baktığımızda, iki satırın da birbirine oranı olan $\frac{-1}{-1}=\frac{1}{1}$ olduğunu görürüz. Yani iki satır da doğrusal bağımlıdır. Bu durumda, $A$ matrisinin rankı 1'dir.

$
rank(A)=1
$

$A|b$ matrisine baktığımızda ise, iki satırın birbirine oranı $\frac{-4}{0}$ olacağından tanımsızdır. Yani iki satır doğrusal bağımsızdır. Bu durumda, $A|b$ matrisinin rankı 2'dir.

$
rank(A|b)=2
$

Görüldüğü gibi, $rank(A)\neq rank(A|b)$ olduğundan, denklem sisteminin çözümü yoktur. Yani denklem sisteminin çözüm kümesi **boşkümedir.**.

<b>Örnek:</b>

$2x+5y=4$

$4x+10y=8$

Denklem sisteminin çözümü var mıdır?

Öncelikle, katsayılar matrisi $A$ ve sabitler matrisi $b$'yi yazalım.

$
A=\begin{bmatrix}
2 & 5 \\
4 & 10
\end{bmatrix}
$

$
b=\begin{bmatrix}
4 \\
8
\end{bmatrix}
$

Artırılmış matris $A|b$ ise şöyle olur:

$
A|b=\begin{bmatrix}
2 & 5 & | & 4 \\
4 & 10 & | & 8
\end{bmatrix}
$

Şimdi, $A$ matrisinin determinantını hesaplayalım.

$
|A|=\begin{vmatrix}
2 & 5 \\
4 & 10
\end{vmatrix}=2\times 10-5\times 4=0
$

Görüldüğü gibi, $|A|=0$ olduğundan, $A$ matrisinin tersi yoktur. Yani $A^{-1}$ mümkün değildir. Bu durumda, denklem sistemi **Cramer olmayan** bir denklem sistemidir.

Bu durumda, denklem sisteminin çözümlü olabilmesi için gerek ve yeter şart olan $rank(A)=rank(A|b)$ eşitliğini kontrol etmemiz gerekir.

Bunun için, $A$ ve $A|b$ matrislerinin ranklarını bulalım.

Bir matrisin rankı, o matrisin doğrusal bağımsız satır sayısını verir. Doğrusal bağımsız satır sayısı ise, o satırların birbirine oranlanamayacağı anlamına gelir.

$A$ matrisine baktığımızda, iki satırın da birbirine oranı olan $\frac{-2}{-4}=\frac{5}{10}$ olduğunu görürüz. Yani iki satır da doğrusal bağımlıdır. Bu durumda, $A$ matrisinin rankı 1'dir.

$
rank(A)=1
$

$A|b$ matrisine baktığımızda ise, iki satırın birbirine oranı $\frac{-2}{-4}=\frac{4}{8}$ olduğunu görürüz. Yani iki satır da doğrusal bağımlıdır. Bu durumda, $A|b$ matrisinin rankı da 1'dir.

$
rank(A|b)=1
$

Görüldüğü gibi, $rank(A)=rank(A|b)$ olduğundan, denklem sisteminin çözümü vardır. Yani denklem sistemi **bağdaşandır**. Ancak bu durumda denklem sisteminin çözüm kümesi **sonsuzdur**.

<b>Örnek:</b>

$2x+5y=3$

$4x-y=2$

Çözümü var mıdır?

### Python İle Çözüm

```python
import numpy as np

# Katsayılar matrisi A ve sabitler matrisi b'yi tanımla
A = np.array([[2, 5], [4, -1]])
b = np.array([3, 2])

# Numpy'nin solve fonksiyonu ile denklem sisteminin çözümünü bul
x = np.linalg.solve(A, b)

# Çözümü yazdır
print(x)
```

## Gauss Yoketme Metodu

Bir lineer denklem sistemine ait artırılmış matrisi satır işlemlerini kullanarak satır eşelon formuna getirmek ve bu sadeleştirilmiş formu kullanarak denklem sistemini çözmek için kullanılan bir yöntemdir. Bu yöntemde, matrisin sol üst köşesinden başlayarak, her sütunda sıfır olmayan ilk elemanı 1 yapmak ve altındaki elemanları sıfırlamak için satır işlemleri uygulanır. Bu şekilde, matrisin sol üst köşesinden sağ alt köşesine doğru bir merdiven şekli oluşturulur. Bu formda, denklem sistemi geriye doğru çözülerek bilinmeyenler bulunur.

<b>Örnek:</b>

$2x+y-2z=10$

$3x+2y+2z=1$

$5x+4y+3z=4$

Lineer deklem sisteminin çözüm kümesini bulunuz.

<b>Çözüm:</b> Adım adım çözelim.

1. Denklem sistemine ait artırılmış matrisi yazalım:

$
\begin{bmatrix}
2 & 1 & -2 & | & 10 \\
3 & 2 & 2 & | & 1 \\
5 & 4 & 3 & | & 4
\end{bmatrix}
$

2. İlk sütunda sıfır olmayan ilk elemanı 1 yapmak için ilk satırı 2'ye bölelim:

$
\begin{bmatrix}
1 & \frac{1}{2} & -1 & | & 5 \\
3 & 2 & 2 & | & 1 \\
5 & 4 & 3 & | & 4
\end{bmatrix}
$

3. Altındaki elemanları sıfırlamak için ilk satırın -3 katını ikinci satıra ve -5 katını üçüncü satıra ekleyelim:

$
\begin{bmatrix}
1 & \frac{1}{2} & -1 & | & 5 \\
0 & \frac{1}{2} & 5 & | & -14 \\
0 & \frac{3}{2} & 8 & | & -21
\end{bmatrix}
$

4. İkinci sütunda sıfır olmayan ilk elemanı 1 yapmak için ikinci satırı 2'ye bölelim:

$
\begin{bmatrix}
1 & \frac{1}{2} & -1 & |& 5 \\
0&1&10&|&-28\\
0&\frac{3}{2}&8&|&-21
\end{bmatrix}
$

5. Altındaki elemanları sıfırlamak için ikinci satırın -$\frac{3}{2}$ katını üçüncü satıra ekleyelim:

$
\begin{bmatrix}
1&\frac{1}{2}&-1&|&5\\
0&1&10&|&-28\\
0&0&-7&|&21
\end{bmatrix}
$

6. Üçüncü sütunda sıfır olmayan ilk elemanı 1 yapmak için üçüncü satırı -$\frac{1}{7}$ ile çarpalım:

$
\begin{bmatrix}
1&\frac{1}{2}&-1&|&5\\
0&1&10&|&-28\\
0&0&1&|&-3
\end{bmatrix}
$

7. Denklem sistemini geriye doğru çözelim:

$z=-3$

$y+10z=-28 \implies y=-28-10(-3) \implies y=2$

$x+\frac{1}{2}y-z=5 \implies x=5-\frac{1}{2}(2)+(-3) \implies x=1$

8. Denklem sisteminin çözüm kümesini yazalım:

$\{(1,2,-3)\}$

### Python İle Çözümü

```python
# Artırılmış matrisi bir liste olarak tanımlayalım
A = [[2, 1, -2, 10],
     [3, 2, 2, 1],
     [5, 4, 3, 4]]

# Matrisin boyutunu bulalım
n = len(A)

# Her sütun için
for i in range(n):
    # Sıfır olmayan ilk elemanı 1 yapmak için satırı bölelim
    pivot = A[i][i]
    for j in range(i, n+1):
        A[i][j] /= pivot
    
    # Altındaki elemanları sıfırlamak için satır işlemleri yapalım
    for k in range(i+1, n):
        factor = A[k][i]
        for j in range(i, n+1):
            A[k][j] -= factor * A[i][j]

# Geriye doğru çözüm yapalım
x = [0] * n # Çözüm kümesini tutacak bir liste oluşturalım
for i in range(n-1, -1, -1): # Son satırdan başlayarak yukarı doğru ilerleyelim
    x[i] = A[i][n] # Sağ taraftaki değeri alalım
    for j in range(i+1, n): # Üstündeki elemanları çıkartalım
        x[i] -= A[i][j] * x[j]

# Çözüm kümesini yazdıralım
print(x)
```
