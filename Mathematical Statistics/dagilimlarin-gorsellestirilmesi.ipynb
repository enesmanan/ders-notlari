```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Ortalama ve kovaryans matrisi
mu = np.array([0, 0])
cov = np.array([[1, 0.5], [0.5, 1]])

rv = multivariate_normal(mean=mu, cov=cov)

# X ve Y koordinatları aralığı
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x
pos[:, :, 1] = y

pdf = rv.pdf(pos)

fig, ax = plt.subplots(facecolor='whitesmoke')
ax.contourf(x, y, pdf, cmap='Blues')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('2 Boyutlu Normal Dağılım')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import warnings
warnings.filterwarnings('ignore')

# ilk grafik: 3D normal dağılım verileri
mean = [0, 0]
cov = [[1, 0.5], [0.5, 1]]
x, y = np.random.multivariate_normal(mean, cov, 500).T
z = np.zeros_like(x)
colors = np.sqrt(x**2 + y**2)
cmap = plt.cm.get_cmap('cool')
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x, y, z, c=colors, cmap=cmap, s=15)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('İki Değişkenli Normal Dağılım')
ax1.view_init(elev=20, azim=45)
ax1.dist = 12
fig.colorbar(cm.ScalarMappable(norm=None, cmap=cmap), shrink=0.5)

# ikinci grafik
mu_x, sigma_x = 0, 2
mu_y, sigma_y = 0, 2
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-((X - mu_x)**2 / (2 * sigma_x**2) + (Y - mu_y)**2 / (2 * sigma_y**2)))
ax2 = fig.add_subplot(122, projection='3d')
surf = ax2.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax2.view_init(elev=20, azim=45)
ax2.dist = 12
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('3 Boyutlu Normal Dağılım')
plt.show()
```
